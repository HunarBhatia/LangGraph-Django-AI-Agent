```python
from langgraph.checkpoint.memory import InMemorySaver
```

```python
from ai.agents import get_document_agent
```

```python
import uuid
config={"configurable":{"user_id":"1","thread_id":uuid.uuid4()}}
checkpointer=InMemorySaver()
agent=get_document_agent(checkpointer=checkpointer)
```

```python
response=agent.invoke(
    {
        "messages":
        [
            {"role":"user" , "content":"Give me a summary of the movie barbie."}
        ]
    },
    config
)
response['messages'][-1].content
```

```python
response=agent.invoke(
    {
        "messages":
        [
            {"role":"user" , "content":"Great. Now, list all documents that i have. After that create a new document with barbie title and add this summary as content"}
        ]
    },
    config
)
response['messages'][-1].content
```
